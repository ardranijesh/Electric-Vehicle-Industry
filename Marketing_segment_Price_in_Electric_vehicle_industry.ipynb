```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
price_change= pd.read_csv('Petrol Price change .csv')
```

```python
price_change
```

```python
plt.figure(figsize=(20, 6))
sns.barplot('DATE', 'PETROL PRICE RS./LIT', data=price_change)
plt.title('Petrol Price change' , fontsize = 16)
plt.ylabel('Petrol Price change')
plt.xlabel('date')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot('YEAR', 'PETROL PRICE RS./LIT', data=price_change)
plt.title('Petrol Price change' , fontsize = 16)
plt.ylabel('Petrol Price change')
plt.xlabel('date')
plt.xticks(rotation=75)
plt.show()
```

```python
data=pd.read_csv("states_data_car.csv")
data.head(20)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
sns.countplot(x=data["road tax"])
plt.title('Count plot of road tax')
plt.xlabel('road tax')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.barplot('state', 'petrol', data=data)
plt.title('Petrol Price in Different States' , fontsize = 16)
plt.ylabel('Petrol Price')
plt.xlabel('State Name')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot('state', 'diesel', data=data)
plt.title('Diesel Price at different states' , fontsize = 16)
plt.ylabel('Diesel Price')
plt.xlabel('State Name')
plt.xticks(rotation=90)
plt.show()
```

# Conclusion

Electric vehicles (EVs) have emerged as a superior choice over traditional vehicles in the face of skyrocketing petrol prices. With the increasing cost of petrol, EVs offer substantial savings on fuel expenses due to their efficiency in utilizing electricity as a fuel source. Moreover, EV owners benefit from stable and predictable energy prices, as electricity costs tend to be less volatile compared to petrol. In addition to cost savings, EVs require lower maintenance and operating costs due to their simplified mechanics and regenerative braking systems. Governments worldwide are also incentivizing the adoption of EVs through subsidies and tax credits. Ultimately, choosing an electric vehicle not only provides financial advantages but also contributes to a cleaner and greener future.
